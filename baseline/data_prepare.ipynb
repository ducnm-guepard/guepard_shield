```python
import pandas as pd
```

```python
df = pd.read_csv("./data/sqli.csv")
```

```python
statements = df[
    df["Query"].str.strip().str.match(r"^(select|update|insert|delete|with)\b", case=False)
]
```

```python
statements["Query"] = statements["Query"].str.strip()
```

```python
statements.to_csv('./data/sqli_statement.csv', index=False, header=True)
```
